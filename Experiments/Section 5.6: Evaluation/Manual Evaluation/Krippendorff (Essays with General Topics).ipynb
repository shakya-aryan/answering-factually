## All models

```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt 
import seaborn 
import krippendorff  


data = pd.read_csv('/bigwork/nhwpshaa/pilot experiments/Evaluation for essays on a general topic.csv', header=None)
evaluation = data.to_numpy() #convert data to numpy array
kalpha = krippendorff.alpha(reliability_data=evaluation, level_of_measurement='ordinal')
print("Krippendorff's alpha score: ", kalpha)

annotators = evaluation.shape[0]# number of annotators (rows)
matrix = np.zeros((annotators, annotators))  # initialization

for i in range(annotators):
    for j in range(i, annotators):
        boolArr = evaluation[i, :] == evaluation[j, :] # check if each value matches and store in a boolean array 
        agreements = np.sum(boolArr)  # count agreements
        matrix[i, j] = agreements
        matrix[j, i] = agreements 


plt.figure(figsize=(11, 9))  
seaborn.heatmap(matrix, annot=True, cmap=seaborn.cubehelix_palette(as_cmap=True), vmin=0, 
            xticklabels=[f'Annotator {i+1}' for i in range(annotators)], 
            yticklabels=[f'Annotator {j+1}' for j in range(annotators)])
plt.title('Annotator Agreement Matrix')  
plt.xlabel('Annotators')  
plt.ylabel('Annotators')  
plt.show()

avg = np.mean(evaluation, axis= 1)
plt.figure(figsize=(10, 8)) 
plt.bar([f'Annotator {i+1}' for i in range(evaluation.shape[0])], avg, color='purple')
plt.title('Average Score by Annotator')
plt.xlabel('Annotator')
plt.ylabel('Average Score')
plt.show()
```
